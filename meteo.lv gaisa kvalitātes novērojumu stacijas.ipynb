```python
import requests as reqs
import json
```

```python
url="https://videscentrs.lvgmc.lv/data/gaiss_stacijas"
gaisa_stacijas=reqs.get(url)
```

```python
stacijas_json=json.loads(gaisa_stacijas.text)
```

Kādu informāciju var atrast šajos datos:

```python
keys=stacijas_json[0].keys()
for key in keys:
    print(key)
```

Šīs atslēgas var izmantot, lai ciklā atlasītu, parādītu informāciju

Kādi no informācijas laukiem var saturēt vairākus ierakstus, kā, piemēram - `mērierīces`, `novērojumi` un `attēli`

```python
for enum, stacija in enumerate(stacijas_json):
    print(enum, stacija["nosaukums"],", ", stacija["adrese"])
    
    for ierice in stacija["merierices"]:
        print("\t",ierice["metode"], ierice["tips"], ierice["raditajs_nosaukums"])
        
    for noverojumi in stacija["noverojumi"]:
        print("\t\t",noverojumi["nosaukums"])
```

```python
for each in stacijas_json:
    if len(each["atteli"])>0:
        print(each["nosaukums"],each["grupa"])
        for file in each["atteli"]:
            print("https://videscentrs.lvgmc.lv"+ file["file_info"]["link"])
```
